```python
#     NumPy   Pyplot
import numpy as np #     . 
import matplotlib.pyplot as plt # 

x=np.linspace(0, 10, 100) #   100   0  10

y=np.sin(x) #     
z=np.cos(x) #     

plt.figure(figsize=(10,5)) #

plt.plot(x,y,label='sin(x)',color='violet') #      
plt.plot(x,z,label='cos(x)',color='grey')

plt.xlabel('x') #  
plt.ylabel('y,z') # Y
plt.title(' sin(x)  cos(x)') # 

plt.legend() #     
plt.show() # 
```

```python
#     NumPy   Pyplot
import numpy as np #     . 
import matplotlib.pyplot as plt # 

#  55  .   1  10
mat=np.random.randint(1, 11, size=(5,5))

print(' ')
print(mat)


plt.imshow(mat, cmap='YlGnBu') #imshow -   , cmap   
plt.colorbar() #
plt.title(' ') #  
#   
for i in range(5):
    for j in range(5):
        if mat[i,j] > 5:
            plt.text(j, i, mat[i,j], ha='center', va='center', color='white')
        else: 'black'

plt.show() #  
```
